```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np 
```

# Matplotlib Basics

```python
x = np.linspace(0, 5, 11)
x

y = x**2
y
```

```python
plt.plot(x, y)
plt.title("y = x^2")
plt.xlabel('X axis')
plt.ylabel("Y axis (x^2)")
#first we pass x-axis, and then we pass y-axis
```

# Subplots in Matplotlin

```python
plt.subplot(2, 2, 1)
plt.xlabel("X-axis")
plt.plot(x, y)
plt.subplot(2, 2, 2)
plt.xlabel("X-axis")
plt.plot(x**2, y)
plt.subplot(2, 2, 3)
plt.plot(x, y)
plt.xlabel("X-axis")
plt.subplot(2, 2, 4)
plt.plot(y, x)
plt.xlabel("X-axis")
```

# Object Oriented Way

```python
fig = plt.figure(figsize = (12, 18), dpi= 100) #this is a canvas
axis1 = fig.add_axes([0.1, 0.1, 1, 1])
axis1.plot(x, y)
axis2 = fig.add_axes([0.3,0.3, 0.3, 0.3])
axis2.plot(x, y)
axis2.plot(x, y)
```

# Types of plot

```python
plt.scatter(x, y)
```

```python
from random import sample
data = sample(range(1, 1000), 100)
plt.hist(data)
#only x is provided
```

```python
plt.boxplot(x)
```

# Saving Plots

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, y)
plt.savefig("basicplot.png")
```
