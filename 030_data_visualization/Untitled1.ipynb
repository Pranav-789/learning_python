```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Distribution Plots


We use distribution plots to visualize the distribution of quantitative data.

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

distplot/histplot
jointplot
pairplot
rugplot

```python
df = sns.load_dataset('tips')
df
#Quantitative datas: total_bill, tip, size
#Categorizable/Qualitative data: sex, smoker, day, time
```

```python
df['size'].unique()
#as there are only 6 unique sizes, therefore size can be refered as Qualitative data
```

```python
sns.displot(df['total_bill'], bins = 20) #depricated, i guess
```

```python
plt.subplot(1, 2, 1)
sns.histplot(df['total_bill'], bins = 20, kde=True)

plt.subplot(1, 2, 2)
sns.histplot(df['tip'], bins = 20, kde=True)
```

# JointPlot

```python
sns.jointplot(x = "total_bill", y = 'tip', data = df, kind='reg')
```

# Pair Plot

```python
sns.pairplot(df, hue='sex', palette='rainbow')
```

# Rug Plot

```python
sns.rugplot(df['total_bill'])
```
